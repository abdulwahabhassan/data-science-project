## Matplotlib example workflow

```python
# 0. Import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("images/sample-plot.png")
plt.show()
```

## Making figures with NumPy arrays

We want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2) # x**2, represent y values derived from values of x raised to the power of 2
plt.show()
```

```python
# Use same data to make a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))  # np.exp(x), represent y values derived from exponential values of x
plt.show()
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
plt.show()
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12
                    }
fig, ax = plt.subplots()
ax.bar(x=nut_butter_prices.keys(), height=nut_butter_prices.values())
ax.set(title="Dan's Nut Butter Store", ylabel="Price ($)")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values())# Horizonta bar graph
plt.show()
```

```python
# Make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
plt.show()
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
plt.show()
```

## Which one to use? (pyplot vs matplotlib object oriented method)
* Use matplotlib oo method for flexibility and more advanced plot
* Otherwise use the pyplot method

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter', x='age', y='chol', c='target')
plt.show()
```

```python
# OO method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax)
# ax.set_xlim([45, 100]) # display only x axis data within the specified limit
plt.show()
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# Customoze the plot
ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target") # * - unpacks the result of scatter.legend_elements()

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--')

plt.show()
```

```python
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

# Add data to ax0
scatter_chol = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels", ylabel="Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter_chol.legend_elements(), title="Target")

# Add a mean line
ax0.axhline(over_50["chol"].mean(), linestyle="--")

# Add data to ax1
scatter_thal = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Thalach")

# Add a legended to ax_100
ax1.legend(*scatter_thal.legend_elements(), title="Target")

# Add a mean line
ax1.axhline(over_50["thalach"].mean(), linestyle="--")

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")

plt.show()
```

```python
# Set the style
plt.style.use('seaborn-v0_8-whitegrid')

## OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(
    x=over_50["age"], 
    y=over_50["chol"], 
    c=over_50["target"],
    cmap="winter" # This changes the color scheme
)

# Customoze the plot
ax.set(title="Heart Disease and Cholesterol Levels", xlabel="Age", ylabel="Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target") # * - unpacks the result of scatter.legend_elements()

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle='--')

plt.show()
```

```python
# Customizing the y and x axis limitations

# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10), sharex=True)

# Add data to ax0
scatter_chol = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"], cmap='winter')

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels", ylabel="Cholesterol")
ax0.set_xlim(50, 80) # Limit x-axis

# Add a legend to ax0
ax0.legend(*scatter_chol.legend_elements(), title="Target")

# Add a mean line
ax0.axhline(over_50["chol"].mean(), linestyle="--")

# Add data to ax1
scatter_thal = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"],cmap='winter')

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Thalach")
ax1.set_xlim(50, 80) # Limit x-axis
ax1.set_ylim(60, 200) # Limit x-axis

# Add a legended to ax_100
ax1.legend(*scatter_thal.legend_elements(), title="Target")

# Add a mean line
ax1.axhline(over_50["thalach"].mean(), linestyle="--")

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")

plt.show()
```
